```python
# Reload the dataset to ensure a clean state
df = pd.read_csv('/content/Algerian_forest_fires_dataset.csv')

 df.drop(['day','month','year', 'Temperature'],axis=1,inplace=True)

 df.columns = df.columns.str.strip()

 if 'Classes' in df.columns:
    df = df.drop('Classes', axis=1)

 for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

 df.dropna(inplace=True)


df = pd.read_csv('/content/Algerian_forest_fires_dataset.csv')

 df.columns = df.columns.str.strip()

 df['Classes'] = np.where(df['Classes'].str.contains('not fire', na=False), 0, 1)


df.drop(['day','month','year', 'Temperature'],axis=1,inplace=True)
if 'Classes' in df.columns and df['Classes'].dtype == 'object':
     df = df.drop('Classes', axis=1)


 for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dropna(inplace=True)


x = df.drop(['Classes'],axis=1)
y = df['Classes']

# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

X_train.shape,X_test.shape , y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
```

```python
ridge_model = Ridge()
lasso_model = Lasso()
elastic_net_model = ElasticNet()
```

```python
ridge_model.fit(X_train, y_train)
lasso_model.fit(X_train, y_train)
elastic_net_model.fit(X_train, y_train)
```

```python
# Make predictions
ridge_pred = ridge_model.predict(X_test)
lasso_pred = lasso_model.predict(X_test)
elastic_net_pred = elastic_net_model.predict(X_test)

ridge_mse = mean_squared_error(y_test, ridge_pred)
ridge_r2 = r2_score(y_test, ridge_pred)

lasso_mse = mean_squared_error(y_test, lasso_pred)
lasso_r2 = r2_score(y_test, lasso_pred)

elastic_net_mse = mean_squared_error(y_test, elastic_net_pred)
elastic_net_r2 = r2_score(y_test, elastic_net_pred)


print("Ridge Regression:")
print(f"  Mean Squared Error: {ridge_mse:.4f}")
print(f"  R-squared: {ridge_r2:.4f}")
print("\nLasso Regression:")
print(f"  Mean Squared Error: {lasso_mse:.4f}")
print(f"  R-squared: {lasso_r2:.4f}")
print("\nElastic Net Regression:")
print(f"  Mean Squared Error: {elastic_net_mse:.4f}")
print(f"  R-squared: {elastic_net_r2:.4f}")
```

```python

ridge_pred_scaled = ridge_model.predict(X_test_scaled)
lasso_pred_scaled = lasso_model.predict(X_test_scaled)
elastic_net_pred_scaled = elastic_net_model.predict(X_test_scaled)


ridge_mse_scaled = mean_squared_error(y_test, ridge_pred_scaled)
ridge_r2_scaled = r2_score(y_test, ridge_pred_scaled)

lasso_mse_scaled = mean_squared_error(y_test, lasso_pred_scaled)
lasso_r2_scaled = r2_score(y_test, lasso_pred_scaled)

elastic_net_mse_scaled = mean_squared_error(y_test, elastic_net_pred_scaled)
elastic_net_r2_scaled = r2_score(y_test, elastic_net_pred_scaled)

print("Ridge Regression (Scaled Data):")
print(f"  Mean Squared Error: {ridge_mse_scaled:.4f}")
print(f"  R-squared: {ridge_r2_scaled:.4f}")
print("\nLasso Regression (Scaled Data):")
print(f"  Mean Squared Error: {lasso_mse_scaled:.4f}")
print(f"  R-squared: {lasso_r2_scaled:.4f}")
print("\nElastic Net Regression (Scaled Data):")
print(f"  Mean Squared Error: {elastic_net_mse_scaled:.4f}")
print(f"  R-squared: {elastic_net_r2_scaled:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
# X = store data of class 1
# y = store data of class 2
#  there are 2 classes means binary classification
```

```python
pd.DataFrame(y)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.predict_proba(X_test)
# The code model.predict_proba(X_test) calculates the probability of each sample in the X_test dataset belonging to each class.

# For a binary classification problem like this one (with 2 classes), the output is an array where each row corresponds to a sample in X_test, and the two columns represent the probabilities:

# The first column is the probability of the sample belonging to class 0.
# The second column is the probability of the sample belonging to class 1.
# The sum of the probabilities in each row will always be 1. This output is useful for understanding the model's confidence in its predictions and can be used to set different thresholds for classification.
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
score
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```

# Grid Search cv


```python
model = LogisticRegression()
```

```python
penalty = ['l1','l2']
c_values = [100,10,1.0,0.1,0.01]
solver = ['liblinear','saga'] # 'newton-cg', 'sag' do not support l1 penalty. 'liblinear' and 'saga' support both l1 and l2. 'elasticnet' is not supported by 'newton-cg' and 'sag'.
```

```python
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
# grid seach cv
from sklearn.model_selection  import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model,param_grid=params,scoring = 'accuracy',cv=cv,n_jobs=-1)
```

```python
print(grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
score
cm = confusion_matrix(y_test,y_pred)
cm
cr = classification_report(y_test,y_pred)
print(cr)
```

# Randomized search cv

```python
from sklearn.model_selection import RandomizedSearchCV

```

```python
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator=model,param_distributions=params,scoring = 'accuracy',cv=cv,n_jobs=-1)
```

```python
randomcv.fit(X_train,y_train)
```

```python
print(randomcv)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
score
cm = confusion_matrix(y_test,y_pred)
cm
cr = classification_report(y_test,y_pred)
print(cr)
```
